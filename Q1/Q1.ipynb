## Q1. What is the distribution of SalePrice in the dataset, and is it skewed or normal?

```python
df = pd.read_csv('./../DataFiles/house-prices-advanced-regression-techniques/train.csv')
df.head()
```

```python
plt.figure(figsize=(9, 8))
sns.histplot(df["SalePrice"], kde=True, bins=100, color='g', alpha=0.4)
```

Answer:
------------
mean = 180921.195890, 
std = 79442.502883

SalePrice follow Normal Distribution, but not symmetrical.
$$\mu = 180921 \\ \sigma = 79442$$

It is visible that the distribution of SalePrice is skewed to the left. This means that the mean is smaller than the median.
