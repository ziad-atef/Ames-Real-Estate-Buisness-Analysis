## Q2. Is there a significant difference in the mean SalePrice between properties with and without a pool (PoolArea > 0 vs. PoolArea = 0)?

```python
df = pd.read_csv(
    './../DataFiles/house-prices-advanced-regression-techniques/train.csv')
df.head()
```

```python
df_with_pool = df[df['PoolArea'] > 0]
df_without_pool = df[df['PoolArea'] == 0]
```

```python
mean_with_pool = df_with_pool['SalePrice'].mean()
mean_without_pool = df_without_pool['SalePrice'].mean()

# get variance
var_with_pool = df_with_pool['SalePrice'].var()
var_without_pool = df_without_pool['SalePrice'].var()

print('Mean with pool: ', mean_with_pool)
print('Variance with pool: ', var_with_pool)

print('Mean without pool: ', mean_without_pool)
print('Variance without pool: ', var_without_pool)
```

```python
ttest_ind(df_with_pool['SalePrice'], df_without_pool['SalePrice'], equal_var=False)
```

Answer:
------------
When comparing means of the samples we found there may be a difference but when testing for statistical significance there is no significant difference in the mean SalePrice between properties with and without a pool.
